```python
os.environ["GROQ_API_KEY"] = os.environ.get('GROQ_API_KEY')
# This line of code is setting an environment variable named "GROQ_API_KEY" to the value retrieved from the existing environment variable
```

```python
# calling the Groq API using the langchain library
from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage
# The above code is importing the necessary modules and classes from the langchain library to initialize a chat model and create messages for the chatbot. The init_chat_model function is used to initialize the chat model, while HumanMessage and SystemMessage are classes that represent different types of messages in the chatbot conversation. HumanMessage typically represents messages from the user, while SystemMessage represents messages from the system or chatbot itself.
```

```python
model = init_chat_model("groq:llama-3.3-70b-versatile")
model
```

```python
message  = [
    SystemMessage("You are a helpful assistant that helps answer questions about the sports."),
    HumanMessage("What is the captain of France Football team?")

]
```

```python
# invoke the model with the message
response = model.invoke(message)
response
```

```python
# streaming response-sometimes the response from the model can be large and may take some time to generate. In such cases, it can be beneficial to stream the response as it is being generated, rather than waiting for the entire response to be ready before displaying it. This allows for a more responsive user experience, as the user can start seeing the output as soon as it is available, rather than having to wait for the entire response to be generated.
for chunk in model.stream(message):
    print(chunk.content,end="",flush=True)

```

Dynamic Prompt 


```python
from langchain_core.prompts import ChatPromptTemplate
translation = ChatPromptTemplate.from_messages(
    [
    ("system", "You are a helpful assistant that helps answer questions about the {sport}. Translate the answer from {source_language} to {target_language}."),
    ("user", "{text}")
    ])



# testing the translation prompt
prompt = translation.invoke({
    "sport": "football",
    "source_language": "English",
    "target_language": "French",
    "text": "Who is the captain of France Football team?"
})       

for chunk in model.stream(prompt):
    print(chunk.content,end="",flush=True)
```
