```python
# @tool decorator

from langchain_core.tools import tool
@tool 
def division(a: float, b: float) -> float:
    """Returns the result of dividing a by b."""
    return a/b
```

```python
print(division.name)
print(division.description)
print(division.args)
```

```python
# async implementation
@tool
async def async_division(a: float, b: float) -> float:
    """Returns the result of dividing a by b."""
    return a/b


```

```python
# Annotating a class with @tool
from typing import List,Annotated

@tool
def multiply_by_max(
    a: Annotated[float, "The number to be multiplied"],
    b: Annotated[list[float], "The list of numbers to find the max from"]
) -> float:
    """Multiplies a by the maximum value in b."""
    return a * max(b)   
```
